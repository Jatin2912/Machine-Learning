```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
dataset.keys()
```

```python
print(dataset.DESCR)
```

```python
dataset.target_names
```

```python
dataset.target
```

```python
dataset.feature_names
```

```python
## Let's prepare the dataset in dataframe
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
df.head()
```

```python
df['target']=dataset.target
```

```python
df.head()
```

```python
# divide our dataset into independent and dependent features
X=df.iloc[:,:-1] ## Independent Feature
y=df.iloc[:,-1]  ## dependent Feature
```

```python
X.head()
```

```python
y
```

```python
# Divide the dataset into train and test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
print("Shape of X_train= ",X_train.shape)
print("Shape of y_train= ",y_train.shape)
print("Shape of X_test= ",X_test.shape)
print("Shape of y_test= ",y_test.shape)
```

## Decision Tree Algorithm Model Training

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treeclassifier=DecisionTreeClassifier(random_state=42)
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

### Post Pruning

```python
## Post Pruning with max_depth=2
treeclassifier1=DecisionTreeClassifier(max_depth=2)
```

```python
treeclassifier1.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier1,filled=True)
```

```python
y_pred=treeclassifier1.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,ConfusionMatrixDisplay
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

## Prepruning

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
hyp_parameters={'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],'max_features':['auto','sqrt','log2']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treeclassifier2=GridSearchCV(treeclassifier,param_grid=hyp_parameters,cv=5,scoring='accuracy')
```

```python
treeclassifier2.fit(X_train,y_train)
```

```python
treeclassifier2.best_params_
```

```python
best_estimator=treeclassifier2.best_estimator_
best_estimator
```

```python
best_estimator.tree_
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(best_estimator,filled=True)
```

```python
y_pred1=treeclassifier2.predict(X_test)
```

```python
y_pred1
```

```python
print(confusion_matrix(y_pred1,y_test))
print(accuracy_score(y_pred1,y_test))
print(classification_report(y_pred1,y_test))
```
